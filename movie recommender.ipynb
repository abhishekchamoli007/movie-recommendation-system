```python
import ast
def genresConvert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return (l)
```

```python
movies['genres'] = movies['genres'].apply(genresConvert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'].apply(genresConvert)
```

```python
movies['keywords']=movies['keywords'].apply(genresConvert)
```

```python
movies.head(2)
```

```python
strr=movies.iloc[0].cast
```

```python
def castCrew(obj):
    lst=ast.literal_eval(strr)
    l=[]
    ctr=0
    for i in lst:
        if ctr<2:
            l.append(i['name'])
            ctr+=1
        
    return l
```

```python
castCrew(strr)
```

```python
movies['cast']=movies['cast'].apply(castCrew)
```

```python
movies.head(2)
```

```python
strr=movies.iloc[0].crew
```

```python
movies['crew']=movies['crew'].apply(castCrew)
```

```python
movies.head(2)
```

```python
def collapse(obj):
    l=[]
    for i in obj:
        l.append(i.replace(' ',''))
    return l
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies['tags']=movies['overview']+movies['keywords']+movies['crew']+movies['cast']
```

```python
movies.head(3)
```

```python
movies.iloc[0].tags
```

```python
new=movies.drop(columns=['overview','genres','cast','crew','keywords'])
```

```python
new.head(2)
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head(1)
```

```python
 new['tags']= new['tags'].apply(lambda x:x.lower())
```

```python
new.head(2)
```

```python
pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

    
```

```python
stem(new.iloc[0].tags)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
new.head(1)
```

```python
pip install sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000, stop_words="english")
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
len(similarity[0])
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
    
    for i in distances[1:6]:
        print (new.iloc[i[0]].title)
```
